```python
#Exercise 1
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import csv
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.manifold import Isomap
import pandas as pd

data=pd.read_csv("data/data_2_3_numbers.csv", sep=' ', header=None)

#own code here
model = Isomap(n_components=2)
projected = model.fit_transform(data)

print(data.shape)
print(projected.shape)
plt.scatter(projected[:, 0], projected[:, 1])
plt.colorbar(ticks=range(10))
plt.clim(-0.5, 9.5);
plt.title('PCA projections')
plt.xlabel('PC 1')
plt.ylabel('PC 2')
print("Maalissa")
```

```python
#Exercise 2
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans

#X, y_true = make_blobs(n_samples=800, centers=4,
#                     cluster_std=0.60, random_state=0)
plt.scatter(projected[:, 0], projected[:, 1], s=50);

kmeans = KMeans(n_clusters=4)
kmeans.fit(projected)
y_kmeans = kmeans.predict(projected)
plt.scatter(projected[:, 0], projected[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
#Centers of the clusters are drawn
```
